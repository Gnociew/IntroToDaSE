```python
import random
import time

#选择排序
def SelectSort(list):
    length = len(list)
    
    if length <= 1:
        return list
    
    for i in range(length):
        min = i
        for j in range(i+1,length):
            if list[j] <list[i]:
                min = j
        if min != i :
            temp = list[i]
            list[i] = list[min]
            list[min] = temp
    
    return list

#归并排序
def merge(lList,rList):
    newList = []
    l = r = 0
    while l < len(lList) and r < len(rList):
        if lList[l] < rList[r]:
            newList.append(lList[l])
            l += 1
        else:
            newList.append(rList[r])
            r += 1
        
    if l == len(lList):
        for i in rList[r:]:
            newList.append(i)
    else:
        for i in lList[l:]:
            newList.append(i)
    return newList

def MergeSort(list):
    length = len(list)

    if length <= 1:
        return list
    
    mid = length // 2
    left = MergeSort(list[:mid])
    right = MergeSort(list[mid:])
    
    return merge(left,right)

n = 100000
#可重复随机数
#list = [random.randint(0,n)for i in range(n)]
#无重复随机数
list1 = random.sample([i for i in range(0,n)],n)
list2 = list1

time_start = time.process_time()
SelectSort(list1)
time_end = time.process_time()
time_sum1 = time_end - time_start

time_start = time.process_time()
MergeSort(list2)
time_end = time.process_time()
time_sum2 = time_end - time_start

print(time_sum1,time_sum2)
```

可以发现归并排序的效率更高的，当数量级足够大的时候其效率远远高于选择排序算法。
归并排序的时间复杂度为O(nlogn)，选择排序的时间复杂度为(n^2)。
